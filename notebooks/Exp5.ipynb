```ruby
def plot_line(fname)
    x_arr = []
    y_arr = []
    File.read(fname).split("\n").each do |str|
        a, b = str.split(" ").map{|s| s.to_f} 
        x_arr.push(a)
        y_arr.push(b)
    end
    p = Plot.new
    p.add(:line, x_arr, y_arr)
    p
end
```
